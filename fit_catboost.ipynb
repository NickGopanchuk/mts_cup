**Обучение CatBoost**

```python
import sklearn.metrics as m
from catboost import CatBoostClassifier, CatBoostRegressor, Pool
from sklearn.model_selection import train_test_split
import pandas as pd
import polars as pl
```

```python
df_train = pd.read_parquet('df_train.parquet')
df_test = pd.read_parquet('df_test.parquet')
```

```python
df_train
```

```python
cat_list = list(df_train.select_dtypes('category'))
```

```python
%%time
x_train, x_test, y_train, y_test = train_test_split(\
    df_train.drop(['user_id', 'is_male', 'age'], axis=1), \
                                                    df_train['is_male'], test_size = 0.25, random_state = 12345,\
                                                    stratify=df_train['is_male'])
```

```python
model = CatBoostClassifier(random_state=12345, task_type="GPU", cat_features=cat_list, n_estimators=10000)
```

```python
%%time 
model.fit(x_train, y_train)
```

```python
print(f'GINI по полу {2 * m.roc_auc_score(y_test, model.predict_proba(x_test)[:,1]) - 1:2.3f}')
```

```python
feature_imp = model.get_feature_importance(prettified=True)
```

```python
feature_imp
```

```python
feat_list = list(feature_imp.loc[feature_imp['Importances'] >=0.005]['Feature Id'])
```

```python
len(feat_list)
```

```python
cat_list = list(x_train[[name for name in feat_list]].select_dtypes('category').columns)
```

```python
model = CatBoostClassifier(random_state=12345, task_type="GPU", cat_features=cat_list, n_estimators=10000)
```

```python
%%time 
model.fit(x_train[[name for name in feat_list]], y_train)
```

```python
print(f'GINI по полу {2 * m.roc_auc_score(y_test, model.predict_proba(x_test[[name for name in feat_list]])[:,1]) - 1:2.3f}')
```

```python
model = CatBoostClassifier(random_state=12345, task_type="GPU", n_estimators=10000,\
                           cat_features=list(df_train[[name for name in feat_list]].select_dtypes('category').columns))
```

```python
model.fit(df_train[[name for name in feat_list]], df_train['is_male'], verbose = 1000)
```

```python
sumbit = pd.DataFrame(df_test['user_id'])
```

```python
sumbit
```

```python
sumbit['is_male'] = model.predict_proba(df_test[[name for name in feat_list]])[:,1]
```

```python
sumbit
```

```python
del x_train, x_test, y_train, y_test
```

### По возврасту

```python
%%time
x_train, x_test, y_train, y_test = train_test_split(\
    df_train.drop(['user_id', 'is_male', 'age'], axis=1), \
                                                    df_train['age'], test_size = 0.25, random_state = 12345,\
                                                   stratify=df_train['is_male'])
```

```python
cat_list = list(df_train.select_dtypes('category'))
```

```python
model = CatBoostClassifier(random_state=12345, task_type="GPU", cat_features=cat_list, n_estimators=10000)
```

```python
%%time 
model.fit(x_train, y_train)
```

```python
print(m.classification_report(y_test, model.predict(x_test), \
                            target_names = ['19-25', '25-34', '35-44', '45-54', '55-65', '65+']))
```

```python
feature_imp = model.get_feature_importance(prettified=True)
```

```python
feature_imp 
```

```python
feat_list = list(feature_imp.loc[feature_imp['Importances'] >=0.005]['Feature Id'])
```

```python
len(feat_list)
```

```python
cat_list = list(x_train[[name for name in feat_list]].select_dtypes('category').columns)
```

```python
model = CatBoostClassifier(random_state=12345, task_type="GPU", cat_features=cat_list, n_estimators=10000)
```

```python
%%time 
model.fit(x_train[[name for name in feat_list]], y_train)
```

```python
print(m.classification_report(y_test, model.predict(x_test[[name for name in feat_list]]), \
                            target_names = ['19-25', '25-34', '35-44', '45-54', '55-65', '65+']))
```
